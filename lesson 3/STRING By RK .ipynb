# sep and End

```python
# Default behavior (uses a space)
print("2026", "07", "20")  
# Output: 2026 07 20

# Changing sep to a hyphen
print("2026", "07", "20", sep="-")  
# Output: 2026-07-20

# Changing sep to nothing (empty string)
print("Py", "thon", sep="")  
# Output: Python
```

```python
# Default behavior (moves to a new line)
print("Hello")
print("World")
# Output:
# Hello
# World
```

```python
# Changing end to a space to keep output on the same line
print("Hello", end=" ")
print("World")
# Output: Hello World
```

```python
# Changing end to a custom punctuation marker
print("Loading completed", end="!!!\n")
# Output: Loading completed!!!
```
